# Get common genes, normalize and scale the sets

```python
all_exprs = [CD14_Monocytes,CD19_B_Cells,regulatory_helper_T,Naive_T_Cells,Memory_T_Cells,
        Natural_Killer_Cells,Naive_Cytotoxic_T_Cells,Cytotoxic_T_Cells,A_frozen,C_frozen,Data6k, Data8k]
```

```python
len(all_exprs)
```

```python
all_exprs[0]
```

```python
common_genes =set(all_exprs[0].index)
```

```python
#sets中的dtaframe index找交集
for i in range(1, len(all_exprs)):
        common_genes = set.intersection(set(all_exprs[i].index),common_genes)
```

```python
len(common_genes)
```

```python
#Release unused object 
del CD14_Monocytes, CD19_B_Cells, Regulatory_T_Cells, Naive_T_Cells, 
del Memory_T_Cells, Helper_T_Cells, Natural_Killer_Cells, Naive_Cytotoxic_T_Cells
del Cytotoxic_T_Cells, regulatory_helper_T, A_frozen, C_frozen, Data6k, Data8k
```

# normalize and scale the sets

```python
sep_point = [0]
for i in range(len(all_exprs)):
    all_exprs[i] = all_exprs[i].loc[common_genes,]
    sep_point.append(all_exprs[i].shape[1])
```

```python
sep_point
```

```python
gene_name = list(all_exprs[0].index)
```

```python
sample_name = []
for i in range(len(all_exprs)):
    all_exprs[i].columns
    sample_name.append(all_exprs[i].columns)
```

```python
sample_name[1]
```

```python
total_exprs = np.array(pd.concat(all_exprs, axis = 1, sort = True), dtype = np.float32)
total_exprs
```

```python
pd.concat(all_exprs, axis = 1, sort = True)
```

```python
total_exprs.shape #(genes_IDx "immune_cell + pbmc cell")
```

```python
np.sum(total_exprs, axis=0, keepdims=True).shape
```

```python
#對每個細胞進行 raw counts * 10000
total_exprs = np.divide(total_exprs, np.sum(total_exprs, axis=0, keepdims=True)) * 10000
total_exprs
```

```python
total_exprs.shape #(genes_IDx "immune_cell + pbmc cell")
```

```python
# log 2 transfromation
total_exprs = np.log2(total_exprs+1)
total_exprs
```

```python
total_exprs.shape
```

```python
# 對每個細胞進行 min max scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler(feature_range = (0, 1))
```

```python
total_exprs = scaler.fit_transform(total_exprs)
```

```python
total_exprs
```

# 去除表現量太高或太低的基因

```python
expr = np.sum(total_exprs, axis=1)
expr
```

```python
expr.shape
```

```python
pass_percentile = np.logical_and(expr >= np.percentile(expr, 1), 
                                 expr <= np.percentile(expr, 99))
```

```python
pass_1_gene_name = np.array(gene_name)[np.where(pass_percentile == 1)].tolist()
```

```python
total_exprs = total_exprs[pass_percentile,]
total_exprs
```

```python
total_exprs.shape #(genes x "training set + pbmc cell")
```

```python
len(pass_1_gene_name)
```

# CV selection

```python
cv = np.std(total_exprs, axis=1) / np.mean(total_exprs, axis=1)
cv
```

```python
# 去除cv太高或太低的基因
```

```python
pass_cv = np.logical_and(cv >= np.percentile(cv, 1), cv <= np.percentile(cv, 99))
```

```python
pass_2_gene_name = np.array(gene_name)[np.where(pass_cv == 1)].tolist()
```

```python
with open ("Filtered_gene_list.csv", "w") as outfile:
    outfile.write("{}, \n".format("Ensembl_id"))
    for i in sorted(pass_2_gene_name):
        outfile.write("{}, \n".format(i))
```
