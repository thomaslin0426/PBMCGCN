```python
import numpy as np
import networkx as nx
import pandas as pd
import os 
```

```python
def is_h5(filename):
        return filename.endswith("h5")    
```

```python
file_list = os.listdir("./Processed_new/")
h5_file = sorted(list(filter(is_h5, file_list)))
```

```python
h5_file
```

```python
new_order_h5_file = []
new_order_h5_file.append(h5_file[0])
new_order_h5_file = new_order_h5_file + h5_file[4:12]  + h5_file[1:4]
new_order_h5_file
del h5_file
```

```python
all_exprs = []
for i in range(len(new_order_h5_file)):
    all_exprs.append(pd.read_hdf(str("./Processed_new/" + new_order_h5_file[i])))
    
```

```python
cell_type = ["CD14_Monocytes", "CD19_B_Cells", "Regulatory_helper_T", "Naive_T_Cells", 
        "Memory_T_Cells", "Natural_Killer_Cells", "Naive_Cytotoxic_T_Cells", 
        "Cytotoxic_T_Cells", "A_frozen", "C_frozen", "Data6k", "Data8k"]

number_label = list(range(0,12))
number_label
```

```python
x = np.array(cell_type)
x_1 = np.array(number_label)

label_info = np.repeat(x, [all_exprs[0].shape[1], all_exprs[1].shape[1],
                           all_exprs[2].shape[1], all_exprs[3].shape[1],
                           all_exprs[4].shape[1], all_exprs[5].shape[1],
                           all_exprs[6].shape[1], all_exprs[7].shape[1],
                           all_exprs[8].shape[1], all_exprs[9].shape[1],
                           all_exprs[10].shape[1], all_exprs[11].shape[1]], 
                       axis = 0)
number_label_info = np.repeat(x_1, [all_exprs[0].shape[1], all_exprs[1].shape[1],
                           all_exprs[2].shape[1], all_exprs[3].shape[1],
                           all_exprs[4].shape[1], all_exprs[5].shape[1],
                           all_exprs[6].shape[1], all_exprs[7].shape[1],
                           all_exprs[8].shape[1], all_exprs[9].shape[1],
                           all_exprs[10].shape[1], all_exprs[11].shape[1]], 
                       axis = 0)
```

```python
label_info = pd.DataFrame(label_info.tolist(), columns = ["Cell_type"])
number_label_info = pd.DataFrame(number_label_info, 
                                 columns = ["Number_label"])
```

```python
label_info = pd.concat([label_info, number_label_info], sort = True, 
                       axis = 1)
label_info
label_info.to_csv("./3.Cell_label.csv")
```

```python
total_exprs = pd.concat(all_exprs, axis = 1, sort = True)
del all_exprs
```

```python
total_exprs
```

```python
convert_table = pd.read_csv("4.intersection_gene.csv", index_col = 0)
```

```python
convert_table
```

```python
filter_total_exprs = total_exprs.filter(convert_table["Gene_stable_ID"], 
                                        axis = 0)
del total_exprs
```

```python
filter_total_exprs
```

```python
for i in range(filter_total_exprs.shape[0]):
    filter_total_exprs.rename(index = {convert_table["Gene_stable_ID"][i] : convert_table["Display_name"][i]}, 
                              inplace = True)
```

```python
filter_total_exprs
```

```python
GGI = pd.read_csv("./5.Interaction_network_in_gene_name.csv", index_col = 0)
```

```python
with open("./5.1.Edge_of_gene_gene_interaction_network.csv", "w") as outfile:
    for i in range(GGI.shape[0]):
        outfile.write("{},{} \n".format(GGI["protein1"][i], 
                                        GGI["protein2"][i]))
```

```python
g = nx.read_adjlist("./5.1.Edge_of_gene_gene_interaction_network.csv",
                    delimiter = ",")
node = list(g.nodes)
print("Number of nodes: " + str(len(node)))
```

```python
filter_node_total_exprs = filter_total_exprs.filter(node, axis = 0)
filter_node_total_exprs
```
